```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

# Mengambil data Set

```python
df = pd.read_csv('https://raw.githubusercontent.com/adliif/dataset/main/gender.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

# Preprocessing Data

```python
df_training = df.drop(['gender'], axis=1)
df_training
```

```python
x = df_training.drop(columns=['distance_nose_to_lip_long'])
y = df_training['distance_nose_to_lip_long']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Logistic Regressiom

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
accuracy_score = model.score(X_test, y_test) *100
print(f"{accuracy_score}%")
```

# Naive Bayes

```python
model_nb = GaussianNB()
model_nb.fit(X_train, y_train)
```

```python
accuracy_score_nb = model_nb.score(X_test, y_test) *100
print(f"{accuracy_score_nb}%")
```

# SVM

```python
model_svm = SVC()
model_svm.fit(X_train, y_train)
```

```python
accuracy_score_svm = model_svm.score(X_test, y_test) * 100
print(f"{accuracy_score_svm}%")
```
